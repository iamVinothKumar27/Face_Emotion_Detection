```python
import tensorflow as tf
import keras
import os
import cv2
import random
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
datadirectory="/Users/t.s.vinoth/Desktop/Emotion_Detection/archive /train"
classes=["0","1","2","3","4","5","6"]
```

```python
import os
import cv2
training_data = []

def create_training_data():
    for i in classes:
        path = os.path.join(datadirectory, i)
        label = classes.index(i)
        for j in os.listdir(path):
            try:
                img_path = os.path.join(path, j)
                img_arr = cv2.imread(img_path)

                if img_arr is None:  # Check if the image was read successfully
                    print(f"Image {img_path} could not be read. Skipping.")
                    continue

                new_arr = cv2.resize(img_arr, (224, 224))
                training_data.append([new_arr, label])
            except Exception as e:
                print(f"Error processing image {j} in class {i}: {e}")

    print(f"Training data creation complete. Total images: {len(training_data)}")

create_training_data()
```

```python
training_data=[]
def create_training_data():
    for i in classes:
        path=os.path.join(datadirectory,i)
        lable=classes.index(i)
        # print(i)
        for j in os.listdir(path):
            try:
                img_arr=cv2.imread(os.path.join(path,j))
                new_arr=cv2.resize(img_arr,(224,224))
                training_data.append([new_arr,lable])
                # print(j)
            except Exception as e:
                print(e)
    print("done")
    print(f"Training data creation complete. Total images: {len(training_data)}")
create_training_data()
```

```python
import random
random.shuffle(training_data)
```

```python
x=[]
y=[]
for m,n in training_data:
    x.append(m)
    y.append(n)
x=np.array(x).reshape(-1,224,224,3)
```

```python
y = np.array(y)
```

```python
x=x/255.0
```

```python
weights_path = "mobilenet_v2_weights_tf_dim_ordering_tf_kernels_1.0_224 (2).h5"
model = tf.keras.applications.MobileNetV2(weights=weights_path)
model.summary()
```

```python
base_input=model.input
base_output=model.layers[-2].output
```

```python
final_output=layers.Dense(128)(base_output)
final_output=layers.Activation('relu')(final_output)
final_output=layers.Dense(64)(final_output)
final_output=layers.Activation('relu')(final_output)
final_output=layers.Dense(7,activation='softmax')(final_output)
```

```python
new_model=keras.Model(inputs=base_input,outputs=final_output)
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
new_model.fit(x, y, epochs=12)
```
